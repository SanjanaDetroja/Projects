# Step 6: Analysis and Visualization

```python
# 1.Find all the weather data when weather is clear

print(data[data['Weather'] == 'Clear'].head())
data[data['Weather'] == 'Clear'].shape
```

```python
# 2.find the number of times when the wind speed was exactly  4km/hr:

print(data[data['Wind Speed_km/h'] == 4].head(2))
data[data['Wind Speed_km/h'] == 4].shape
```

```python
# 3.Rename the column name weather to weather_condition:

data.rename(columns = {'Weather':'Weather_condition'},inplace = True )
data.columns
```

```python
# 4.what is the mean visibility of given data_set

data['Visibility_km'].mean()
```

```python
# 5.where visibility is >24 and weather = fog
data[data['Visibility_km']>24 & (data['Weather_condition']=="Fog")].shape
```

```python
# 6. converting into datatime variable
data['Date/Time'] = pd.to_datetime(data['Date/Time'])
data.set_index('Date/Time', inplace=True)
```

```python
# 7. Visibility_km >20
data[data["Visibility_km"] > 20]
```

```python
# 8.where wind speed >24 and visibility is 25
data[data["Wind Speed_km/h"] > 24 & (data["Visibility_km"] == 25)].shape
```

```python
# 9.checking for outliers by box plot

data['Rel Hum_%'].plot(kind='box', figsize = (12,6), fontsize=12)
```

```python
# 10.Monthly Average Temperature Trends

monthly_avg = data['Temp_C'].resample('M').mean()

# Plotting the temperature trends
plt.figure(figsize=(12, 6))
plt.plot(monthly_avg.index, monthly_avg, marker='o', linestyle='-', color='b')
plt.title('Monthly Average Temperature Trends')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.show()
```

```python
# 11.Plotting wind speed trends

fig,ax = plt.subplots(figsize=(18, 6))
ax.plot(data.index, data['Wind Speed_km/h'], marker='*', linestyle='-', color='g')
plt.title('Wind Speed Trends')
plt.xlabel('Date')
plt.ylabel('Wind Speed (m/s)')
plt.grid(True)

max_wind = data['Wind Speed_km/h'].max()
max_wind_index = data['Wind Speed_km/h'].idxmax()

ax.annotate(f'Max Wind: {max_wind}km/h', xy=(max_wind_index, max_wind),
            xytext=(max_wind_index, max_wind +7),
            arrowprops=dict(facecolor='red', shrink=0.01),
           )

plt.show()
```

```python
# 12.Scatter plot for Visibility and Relative Humidity
plt.scatter(data['Visibility_km'], data['Rel Hum_%'], color='red')
plt.title('Scatter Plot: Visibility vs Relative Humidity')
plt.xlabel('Visibility (km)')
plt.ylabel('Relative Humidity (%)')
plt.show()
```

```python
# 13.weather counts

weather_counts = data['Weather_condition'].value_counts()
plt.figure(figsize = (10,22))
plt.barh(weather_counts.index, weather_counts, color=['yellow', 'gray', 'blue', 'green'])
plt.xlabel('Weather Condition')
plt.ylabel('Count')
plt.title('Weather Distribution')
plt.xticks(rotation=45, ha='right')

plt.show()
```

```python
# 14.Creating subplots for the maximum temperature point

fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(data['Temp_C'], label='Temperature (C)')
plt.suptitle('Temperature Over Time')
ax.set_xlabel('Date/Time')
ax.set_ylabel('Temperature (C)')

max_temp = data['Temp_C'].max()
max_temp_index = data['Temp_C'].idxmax()

ax.annotate(f'Max Temp: {max_temp}°C', xy=(max_temp_index, max_temp),
            xytext=(max_temp_index, max_temp +5),
            arrowprops=dict(facecolor='red', shrink=0.01),
            )

ax.legend()
plt.show()
```

```python
# 15.plotting temperature threshold for extreme events

temperature_threshold = 30  # we know max is 33 considering thershold as 30

extreme_events = data[data['Temp_C'] > temperature_threshold]

# Plotting extreme temperature events
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Temp_C'], label='Temperature')
plt.scatter(extreme_events.index, extreme_events['Temp_C'], color='r', label='Extreme Events')
plt.title('Extreme Temperature Events')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.grid(True)
plt.show()
```
